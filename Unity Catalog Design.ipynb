### Unity Catalog Design 
name -> medilakehealth_catalog

**Disclaimer** : In our free Databricks environment, we follow the Unity Catalog structure by creating a catalog and schema, then a volume, where we upload the necessary files for our projects

##### First we list the existing catalogs.

```sql
%sql 

ShOW CATALOGS;
```

**Let's create the medilakehealth_catalog**

```sql
%sql

CREATE CATALOG medilakehealth_catalog;

COMMENT ON CATALOG medilakehealth_catalog IS'we don t have extrnal location';
```

###### Here we check the current catalog is in use.

```sql
%sql

 SELECT current_catalog();
```

###### We need to select our catalog medilakehealth_catalog for use.

```sql
%sql 

USE CATALOG medilakehealth_catalog;
     
     SELECT current_catalog();
```

###### Get more info about the medilakehealth_catalog.

```sql
%sql 

DESCRIBE CATALOG EXTENDED medilakehealth_catalog;
```

##### We create all schemas under the catalog medilakehealth_catalog.

```sql
%sql


CREATE SCHEMA IF NOT EXISTS Bronz;

CREATE SCHEMA IF NOT EXISTS Silver;

CREATE SCHEMA IF NOT EXISTS Gold;

CREATE SCHEMA IF NOT EXISTS landing;
```

##### List the schemas we created.

```sql
%sql 
SHOW SCHEMAS;
```

```sql
%sql 
USE SCHEMA landing;
     
CREATE VOLUME If NOT EXISTS medilakehealth_catalog.landing.operational_data;
```
